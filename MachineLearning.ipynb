# Data Preprocessing

## Impute Missing Values

```python
missingValues = df.isnull().sum()
missingPercentage = (missingValues / len(df)) * 100
dataTypes = df.dtypes
missingData = pd.DataFrame({'Amount of Missing Values': missingValues, 'Percentage of Missing Values': missingPercentage, 'Data Type': dataTypes})
missingData = missingData.sort_values(by=['Percentage of Missing Values'], ascending=False)
missingData.style
```

Kolom **type_of_mean_plan**, **required_car_parking_space**, dan **avg_price_per_room** secara beurutan mempunyai missing values sebesar 4,5%, 3.5%, dan 2.5% dari keseluruhan dataset. Kolom-kolom ini kita bisa handle dengan cara impute mode/median/mean dari data yang ada.

```python
df['type_of_meal_plan'].value_counts()
```

Ternyata missing values pada kolom **type_of_meal_plan** dilabel dengan **Not Selected**.

```python
print(df['type_of_meal_plan'].mode().sum())
```

Setelah dicari tahu lebih lanjut, modus dari kolom **type_of_meal_plan** adalah **Meal Plan 1**.

```python
df['type_of_meal_plan'] = df['type_of_meal_plan'].replace('Not Selected', 'Meal Plan 1')
modus = df['type_of_meal_plan'].mode()[0]
df['type_of_meal_plan'].fillna(modus, inplace=True)
df.head(10)
```

```python
df['type_of_meal_plan'].value_counts()
```

Terakhir, kita impute **Meal Plan 1** yang merupakan modus dari kolom **type_of_meal_plan** ke semua nilai yang **Not Selected** dan **NaN**.

```python
df['required_car_parking_space'].value_counts()
```

```python
print(df['required_car_parking_space'].mode().sum())
```

```python
df['required_car_parking_space'].fillna(df['required_car_parking_space'].mode()[0], inplace=True)
df.head(10)
```

Di sini, kita mengisi kolom **required_car_parking_space** dengan modusnya. Mengapa? Karena walaupun nilai-nilai dalam kolom tersebut direpresentasikan dengan angka, namun angka-angka tersebut bersifat sebagai kategori biner (angka biner). Jadi, tidak masuk akal jika kita mengisi kolom tersebut dengan median/mean, karena keseluruhan kolom tersebut hanya ingin menyatakan apakah pelanggan membutuhkan (1.0) atau tidak membutuhkan (0.0) tempat parkir.

```python
df['avg_price_per_room'].value_counts
```

```python
df['avg_price_per_room'].fillna(df['avg_price_per_room'].median(), inplace=True)
df.head(10)
```

Untuk kolom **avg_price_per_room**, kita dapat mengisi dengan median dari keseluruhan nilai yang ada dalam kolom tersebut.

```python
missingValues = df.isnull().sum()
missingPercentage = (missingValues / len(df)) * 100
dataTypes = df.dtypes
missingData = pd.DataFrame({'Amount of Missing Values': missingValues, 'Percentage of Missing Values': missingPercentage, 'Data Type': dataTypes})
missingData = missingData.sort_values(by=['Percentage of Missing Values'], ascending=False)
missingData.style
```

```python
oneHotEnc = OneHotEncoder(sparse_output=False, drop=None)
labelEnc = LabelEncoder()
```

```python
df['booking_status'] = labelEnc.fit_transform(df['booking_status'])
df['arrival_year'] = labelEnc.fit_transform(df['arrival_year'])
```

Di sini, kita menggunakan **label encoder** karena nilai dalam kolom **booking_status** dan **arrival_year** hanya ada dua nilai yang berbeda.

```python
categorical_cols = ['type_of_meal_plan', 'room_type_reserved', 'market_segment_type']

encoded = oneHotEnc.fit_transform(df[categorical_cols])

encoded_df = pd.DataFrame(encoded, columns=oneHotEnc.get_feature_names_out(categorical_cols))

df = df.drop(columns=categorical_cols)
df = pd.concat([df.reset_index(drop=True), encoded_df], axis=1)
```

Sedangkan untuk kolom **type_of_meal_plan**, **room_type_reserved**, dan **market_segment_type**, kita dapat menggunakan karena nilai masing-masing kolom non-ordinal.

# Data Splitting

```python
x = df.drop(columns=['booking_status'])
y = df['booking_status']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape, x_test.shape)
```

Di sini, kita melakukan splitting data **80% untuk training** dan **20% untuk testing**, di mana target variabel kita adalah kolom **booking_status**.

# Modelling

## Random Forest Model

```python
RF_class = RandomForestClassifier()
RF_class.fit(x_train, y_train)
```

```python
y_pred_RF = RF_class.predict(x_test)

print("Classification Report Random Forest:\n", classification_report(y_test, y_pred_RF, target_names= ['Canceled','Not_Canceled']))
```

## XGBoost Model

```python
XGB_class = XGBClassifier()
XGB_class.fit(x_train, y_train)
y_pred_XGB = XGB_class.predict(x_test)
```

```python
print("Classification Report XGBoost:\n", classification_report(y_test, y_pred_XGB, target_names= ['Canceled','Not_Canceled']))
```
